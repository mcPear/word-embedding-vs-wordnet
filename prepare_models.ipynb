## Filter models, accept only letters

```python
def filter_alpha(vectors, txt_file, mag_file):
    f = open(txt_file, "w")
    f.write('{} {}\n'.format(len(vectors), 100))
    for key,vec in tqdm(vectors):
        #vectors.close()
        if(key.isalpha()):
            f.write('{} {}\n'.format(key, ' '.join(map(str, vec))))
    f.close()
    
    !python3.6 -m pymagnitude.converter -i {modelsDir+txt_file} -o {modelsDir+mag_file}
```

```python
w2v_filtered_file_txt = "skip_gram_v100m8.w2v.filtered.txt"
w2v_filtered_file_mag = "skip_gram_v100m8.w2v.filtered.magnitude"

ft_filtered_file_txt = "kgr10.plain.skipgram.dim100.neg10.filtered.txt"
ft_filtered_file_mag = "kgr10.plain.skipgram.dim100.neg10.filtered.magnitude"

filter_alpha(w2v_vectors, w2v_filtered_file_txt, w2v_filtered_file_mag)
filter_alpha(ft_vectors, ft_filtered_file_txt, ft_filtered_file_mag)
```
